```python
def get_location(ip):
    ip_address = ip
    response = requests.get(f'https://ipapi.co/{ip_address}/json/').json()
    location_data = {
        "ip": ip_address,
        "city": response.get("city"),
        "region": response.get("region"),
        "country": response.get("country_name")
    }
    return location_data
```

```python
# given a json file with a list of ip addresses, return a list of dictionaries with the ip address and location data
def get_locations_from_file(filename):
    with open(filename, 'r') as f:
        ping_list = json.load(f)
    locations = []
    for ping in ping_list:
        print(ping['dst_addr'])
        locations.append(get_location(ping['dst_addr']))
    return locations

destination = get_locations_from_file('ping.json')
```

```python
import geopandas as gpd
from geopy.geocoders import Nominatim
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString
import pandas as pd

# Function to get the coordinates of a country
def get_coordinates(country):
    geolocator = Nominatim(user_agent="my_geocoder")
    location = geolocator.geocode(country)
    if location:
        return Point(location.longitude, location.latitude)
    else:
        print(f"Coordinates not found for {country}")
        return None

# Function to plot countries and connect them with a line
def plot_countries(source_countries, destination_countries, latencies):
    # Create empty GeoDataFrame for points and lines
    gdf_all = gpd.GeoDataFrame(geometry=[])

    for source_country, destination_country, latency in zip(source_countries, destination_countries, latencies):
        # Get coordinates for source and destination
        source_coords = get_coordinates(source_country)
        destination_coords = get_coordinates(destination_country)

        if source_coords and destination_coords:
            # Create GeoDataFrame with points for source and destination
            gdf_points = gpd.GeoDataFrame(geometry=[source_coords, destination_coords])

            # Create GeoDataFrame with a line connecting source and destination
            gdf_line = gpd.GeoDataFrame(geometry=[LineString([source_coords, destination_coords])])

            # Add latency as a column to the line GeoDataFrame
            gdf_line['latency'] = str(latency)

            # Concatenate GeoDataFrames
            gdf_all = pd.concat([gdf_all, gdf_points, gdf_line], ignore_index=True)

    # Plot the world map
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
    ax = world.plot(figsize=(10, 6), color='lightgray')

    # Plot all source and destination points
    gdf_all[gdf_all['geometry'].type == 'Point'].plot(ax=ax, color='red', marker='o', markersize=100)

    # Plot all lines connecting source and destination, annotated with latencies
    for i, row in gdf_all[gdf_all['geometry'].type == 'LineString'].iterrows():
        line = row['geometry']
        latency = row['latency']
        ax.annotate(latency, xy=(line.centroid.x, line.centroid.y), color='black', fontsize=8, ha='center')

    gdf_all[gdf_all['geometry'].type == 'LineString'].plot(ax=ax, color='blue', linewidth=2)

    # Show the plot
    plt.show()

# Example usage with lists of source and destination countries and latencies
source_countries = []
for s in source:
    source_countries.append(s['city'])
destination_countries = []
for d in destination:
    destination_countries.append(d['city'])

latencies = latency
plot_countries(source_countries, destination_countries, latencies)
```

# Traceroute

```python
import json

# get all destination cities from the json file
def get_trace():
    with open('trace.json', 'r') as f:
        ping_list = json.load(f)
    destinations = []
    for ping in ping_list:
        dest = []
        for result in ping['result']:
            if result.get('result') and result['result'][0].get('x'):
                continue
            if result.get('result') and result['result'][0].get('from'):
                dest.append(result['result'][0]['from'])
        destinations.append(dest)
    return destinations

destinations = get_trace()
destinations
```

```python
# for each ip in the destinations list get the location data
trace = []
for dest in destinations:
    t = []
    for i in dest:
        t.append(get_location(i))
    trace.append(t)

trace
        
```

```python
cities = []
for t in trace:
    city_list = []
    for i in t:
        city_list.append(i['city'])
    cities.append(city_list)



# if cities is None remove it
for c in cities:
    for i in c:
        if i == None or i == [] or i == 'None':
            c.remove(i)

# if the list of cities is empty, remove it
cities = [x for x in cities if x != []]
cities
```

```python
import geopandas as gpd
from geopy.geocoders import Nominatim
import matplotlib.pyplot as plt

def plot_cities_on_map(city_list):
    # Initialize geocoder
    geolocator = Nominatim(user_agent="city_plotter")

    # Get coordinates for each city
    coordinates = []
    for city in city_list:
        location = geolocator.geocode(city)
        if location:
            coordinates.append({'City': city, 'Latitude': location.latitude, 'Longitude': location.longitude})

    # Create a GeoDataFrame
    gdf = gpd.GeoDataFrame(coordinates, geometry=gpd.points_from_xy([coord['Longitude'] for coord in coordinates],
                                                                   [coord['Latitude'] for coord in coordinates]))

    # Load world map from geopandas datasets
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

    # Plot the GeoDataFrame on the world map
    fig, ax = plt.subplots(figsize=(10, 6))
    world.plot(ax=ax, color='lightgrey')

    # Calculate plot limits dynamically
    min_longitude, max_longitude = gdf.geometry.x.min(), gdf.geometry.x.max()
    min_latitude, max_latitude = gdf.geometry.y.min(), gdf.geometry.y.max()

    # Add padding for better visualization
    pad_factor = 0.1
    pad_longitude = (max_longitude - min_longitude) * pad_factor
    pad_latitude = (max_latitude - min_latitude) * pad_factor

    ax.set_xlim(min_longitude - pad_longitude, max_longitude + pad_longitude)
    ax.set_ylim(min_latitude - pad_latitude, max_latitude + pad_latitude)

        
    gdf.plot(ax=ax, marker='o', color='red', markersize=50)

    # Annotate each point with the city name
    for x, y, label in zip(gdf.geometry.x, gdf.geometry.y, gdf['City']):
        ax.text(x, y, label, fontsize=8)

    # Connect cities with lines
    for i in range(len(gdf) - 1):
        x_coords = [gdf.geometry.x.iloc[i], gdf.geometry.x.iloc[i + 1]]
        y_coords = [gdf.geometry.y.iloc[i], gdf.geometry.y.iloc[i + 1]]
        ax.plot(x_coords, y_coords, color='blue', linestyle='--')

    # Set plot title
    ax.set_title('Cities on World Map with Connections')

    # Show the plot
    plt.show()

for c in cities: 
    # remove none
    for i in c:
        if i == None or i == [] or i == 'None':
            c.remove(i)
    # plot the cities
    print(c)
    if c != []:
        plot_cities_on_map(c)
```
